```python
csv_path = "/content/drive/MyDrive/Deep_learning/Olympia_dataset/AI_modified_20220809.csv"
img_folder = "/content/drive/MyDrive/Deep_learning/Olympia_dataset/dataset"

#Open_csv
with codecs.open(csv_path, "r", "Shift-JIS", "ignore") as file:
        df = pd.read_csv(file, index_col=None, header=0)
df.iloc[:, 15:]
```

```python
# 眼症なし群を抜き出す
non_GO = df[df["眼症なし"]==1].loc[:,"写真No."].tolist()

# 中等症以上群 + ステロイドあり群
treatment_req_GO = df[df["中等症以上治療なし"]==1].loc[:,"写真No."].tolist() +  df[df["ステロイドあり"]==1].loc[:,"写真No."].tolist()
treatment_req_GO = sorted(treatment_req_GO)

#それぞれの症例数
print(f"non_GO: {len(non_GO)} cases")
print(f"treatment_req_GO: {len(treatment_req_GO)} cases")
```
